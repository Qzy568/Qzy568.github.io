```python
def extract_mp4_filenames(file_path):
    # Open the text file
    with open(file_path, 'r') as file:
        content = file.read()

    # Split the content into words
    words = content.split()

    # Filter out and print the .mp4 filenames
    mp4_files = [word for word in words if word.endswith('.mp4')]
    for file in mp4_files:
        print(file)

# Specify the path to your text file
file_path = '1.txt'
extract_mp4_filenames(file_path)
```
